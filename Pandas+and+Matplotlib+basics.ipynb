```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Create  some test data
df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))

#Explore data and descriptive stats
df.head(10)
```

```python
df.describe()
```

```python
colors = np.where(df.D > 50, 'r', 'k')
df.plot(kind='scatter', x='A', y='B', s = df.C, c =colors)
plt.xlabel('Column A')
plt.ylabel('Column B')
plt.title('Graph of column A v/s column B')
#plt.legend()
plt.show()
```
